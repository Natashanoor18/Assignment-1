```python
from random import randrange

# ---------------- DISPLAY BOARD ----------------
def display_board(board):
    print("+-------+-------+-------+")
    for row in board:
        print("|   {}   |   {}   |   {}   |".format(row[0], row[1], row[2]))
        print("+-------+-------+-------+")


# ---------------- ENTER USER MOVE ----------------
def enter_move(board):
    while True:
        move = input("Enter your move: ")

        if not move.isdigit():
            continue

        move = int(move)

        if move < 1 or move > 9:
            print("Invalid move, choose 1-9")
            continue

        row = (move - 1) // 3
        col = (move - 1) % 3

        if board[row][col] in ['O', 'X']:
            print("Square already taken")
            continue

        board[row][col] = 'O'
        break


# ---------------- FREE FIELDS LIST ----------------
def make_list_of_free_fields(board):
    free = []
    for r in range(3):
        for c in range(3):
            if board[r][c] not in ['O', 'X']:
                free.append((r, c))
    return free


# ---------------- COMPUTER MOVE ----------------
def computer_move(board):
    free = make_list_of_free_fields(board)
    if free:
        r, c = free[randrange(len(free))]
        board[r][c] = 'X'


# ---------------- CHECK VICTORY ----------------
def victory_for(board, sign):
    for row in board:
        if row.count(sign) == 3:
            return True

    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] == sign:
            return True

    if board[0][0] == board[1][1] == board[2][2] == sign:
        return True
    if board[0][2] == board[1][1] == board[2][0] == sign:
        return True

    return False


# ---------------- MAIN GAME ----------------
board = [
    [1, 2, 3],
    [4, 'X', 6],
    [7, 8, 9]
]

display_board(board)

while True:
    enter_move(board)
    display_board(board)

    if victory_for(board, 'O'):
        print("You won!")
        break

    if not make_list_of_free_fields(board):
        print("It's a draw!")
        break

    computer_move(board)
    display_board(board)

    if victory_for(board, 'X'):
        print("Computer won!")
        break
```
